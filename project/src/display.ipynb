```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

hr = pd.read_csv("rocco1.csv")


plt.rcParams["figure.figsize"] = (13,6)
hb = np.array(hr.HR)
t = np.array(hr.Time)
plt.grid()
plt.plot(t,hb, 'r')

plt.xlabel("Time (s)")
plt.ylabel("Heart rate")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

hr = pd.read_csv("tommi1.csv")


plt.rcParams["figure.figsize"] = (13,6)
hb = np.array(hr.HR)
t = np.array([i + 1 for i in range(len(hb))])
plt.grid()
plt.plot(t,hb, 'r')

plt.xlabel("Time (s)")
plt.ylabel("Heart rate")
plt.show()
```
