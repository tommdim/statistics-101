## **Checkmate: one beat at a time.** 
#### *Estimating the position evaluation from the heart rate of players.*



Abstract
----------

*The project aims at finding a correlation between the current position evaluation in a chess match and the heartbeat of the players, we will repeat the experiments using different time controls.*

----------


Main research aim & framework
----------

*Describe the main goal(s) of your project, why you are interested in it, and how you came up with this idea. If there are also secondary goals, list them too.*

The main goal of our project is to investigate the correlation of the heart  rate of a chess player with various features of a match, we expect the heart rate to increase when a player has a short time remaining, we expect an higher heart rate in shorter time controls, still we are not sure how the heart  rate will change in the same way after a blunder or after a brilliant move, in the same way we don't know if the person winning will have an higher rate (respect to his average rate) than the person losing, especially because the formal evaluation of a position isn't disclosed to the players. We plan to group players whose rating differs by at most 200 elo points.

In our group we all know how to play chess, but some of us ci so rimasti sotto. One of them particularly was watching a [tournament](https://www.youtube.com/watch?v=tXlX5dznuPA&t=15078s&ab_channel=Chess.com) where the heart rate of the players was tracked and displayed on screen and thought this could be an interesting experiment to carry out. 

Some of the secondary goals we are aiming to achieve are the prediction of different features using supervised and unsupervised learning, we don't know yet precisily the feature that we will pick, for now.

The only paper we could find to back up our idea (kind of) is the following: Leone, M. J., Petroni, A., Fernandez Slezak, D., & Sigman, M. (2012). *The tell-tale heart: heart rate fluctuations index objective and subjective events during a game of chess*.

----------


Data source(s) 
----------
*Describe the software/hardware/tools/platforms/apps/devices/etc you are planning to use in order to get the data you need.*

We are planning to use a simple chessboard (we are considering if playing online could be more neutral), and to track heart rates using heart rate bands and pulse oximiteters that will be kindly lended by our sporty friends.

We'll consider later on which application to use for storing data.

----------

Data collection
----------
*Explain how you are planning to actually collect the data.*

We are planning to track the heart rate of the partecipants for a bit to estimate their average heart rate, then we are planning to play some chess while our heart rates will be monitored, in the meantime the position evaluation will be written down by someone who isn't actually playing.


*How many data-points do you (realistically) hope to collect?*

We plan to collect as much data as possible, taken from, hopefully, around 50 matches.

*Do you foresee any difficulty in the data collection process?*

Our main concern is that the heartbeat rate would be a too volatile paramether for estimating the position evaluation. 


----------

Methods
----------
The first analysis will be the description of the collected dataset, with plots and numerical statistics.
Then we'll try to find a good density estimation for the data (we can try to use the Gaussian model and MoG), we'll try to find the  empirical CDF and compare it with the parametric model. 
Other methods that we would like to use are clustering, with k-means and using the Elbow method (we'll see during the process if it will give us good results for data classification) and parametric bootstrap for finding the pdf. 

For the last part of the project we'll do some sample splitting and we'll use an evaluation method (like cross validation) for training and testing the dataset; then, using supervised and unsupervised learning, we'll find a prediction for the position evaluation, given the heart rate.
