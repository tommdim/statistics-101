Main research aim & framework
----------
*Describe the main goal(s) of your project, why you are interested in it, 
and how you came up with this idea. If there are also secondary goals, list them too.

The main goal of our project is to investigate the correlation of the heartrate of a chess player with various features of a match, we expect the heart rate to increase when a player has a short time remaining, we expect an higher heart rate in shorter time controls, still we are not sure how the heartrate will change in the same way after a blunder or after a brilliant move, in the same way we don't know if the person winning will have an higher rate (respect to his average rate) than the person losing, especially because the formal evaluation of a position isn't disclosed to the players. We plan to group players whose rating differs by at most 200 elo points.

In our group we all know how to play chess, but some of us ci so rimasti sotto. One of them particularly was watching a [tournament](https://www.youtube.com/watch?v=tXlX5dznuPA&t=15078s&ab_channel=Chess.com) where the heartrate of the players was tracked and displayed on screen and thought this could be an interesting experiment to carry out.
